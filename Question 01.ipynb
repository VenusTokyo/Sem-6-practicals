## Q1) Perform elementary mathematical operations in Octave/MATLAB/R like addition, multiplication, division and exponentiation.

```python
def main():
    ans="Y"
    while(ans=="y" or ans=="Y"):
        print("\n\t-_-_-_-_-_-_-_-_-_-MENU-_-_-_-_-_-_-_-_-_-\n")
        print("1)Addition\n2)Subtraction\n3)Multiplication\n4)Division\n5)Exponentiation")
        ch=int(input("\n\tEnter an option: "))
        if(ch==1):
            x=float(input("Enter Operand 1: "))
            y=float(input("Enter Operand 2: "))
            print("Sum =",x+y)
        elif(ch==2):
            x=float(input("Enter Operand 1: "))
            y=float(input("Enter Operand 2: "))
            print("Difference =",x-y)
        elif(ch==3):
            x=float(input("Enter Operand 1: "))
            y=float(input("Enter Operand 2: "))
            print("Product =",x*y)
        elif(ch==4):
            x=float(input("Enter Dividend: "))
            y=float(input("Enter Divisor: "))
            while(y==0):
                print("Error: Divisor can't be zero! Re-enter value")
                y=float(input("Enter Divisor: "))
            print("Quotient =",x//y,"Remainder =",x%y)
        elif(ch==5):
            x=float(input("Enter base: "))
            y=float(input("Enter power: "))
            print("Result =",x**y)
        else:
            print("Enter valid option")
        
        ans=input("Enter do you want to try again? {Y/N}: ")
        print("\n-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*\n")
        
main()
```
